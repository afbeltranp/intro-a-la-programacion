## Exercise 2: Shopping Cart Calculator

### Task
Create a program that calculates the final total for a shopping cart based on three user inputs. The inputs will be provided one per line:  

1. **Subtotal**: The total cost of the items before any discounts or taxes. When you prompt the user for this input, it will be entered as a float. 
2. **Percent Discount**: The percentage discount to apply to the subtotal. When you prompt the user for this input, it will be entered as a float (e.g., 20 for 20%).  
3. **Tax Percentage**: The percentage tax to apply after the discount. When you prompt the user for this input, it will be entered as a float (e.g., 7.5 for 7.5%).  

The program should perform the following calculations in order:  
1. **Apply the discount first**:  
   - Convert the discount percentage to a fraction by dividing it by 100.  
   - Multiply the fraction by the subtotal to find the discount amount.  
   - Subtract the discount amount from the subtotal to get the discounted price.  

2. **Apply the tax next**:  
   - Convert the tax percentage to a fraction by dividing it by 100.  
   - Multiply the tax fraction by the discounted price to find the tax amount.  
   - Add the tax amount to the discounted price to get the final total.
