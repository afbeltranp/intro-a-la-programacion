# Tutorial de Google Colab

## ¿Qué Es un Cuaderno de Jupyter?

**Pista:** Es lo que acabas de descargar y subir.

Un Cuaderno de Jupyter es un entorno interactivo donde puedes escribir y ejecutar código Python junto con texto, explicaciones y resultados en el mismo documento.

Cuando se usa con Google Colab, un Cuaderno de Jupyter vive en tu Google Drive y te da acceso a un entorno de Python desde casi cualquier lugar.

## Celdas del Cuaderno
Un Cuaderno de Jupyter está compuesto por celdas. Hay dos tipos principales de celdas que usarás.

### Celdas de Código
- Se usan para escribir y ejecutar código Python  
- Se ejecutan usando el botón Ejecutar o Shift + Enter  
- El resultado aparece directamente debajo de la celda  

### Celdas de Texto Formateado
- Se usan para escribir texto, explicaciones o notas  
- No ejecutan código  
- Son útiles para documentar lo que hace tu código.

**Esta celda y todas las que se ven como ella son Celdas de Texto Formateado.** Puedes agregar celdas haciendo clic en +Código o +Texto arriba.

## El Orden de Ejecución Importa ‼️

Los Cuadernos de Jupyter no se ejecutan automáticamente de arriba a abajo. Las celdas se ejecutan solo cuando tú las ejecutas.

Esto significa:
- Las variables pueden existir aunque su celda aparezca más adelante en el cuaderno  
- Ejecutar celdas fuera de orden puede causar errores o resultados confusos  
- Reiniciar el kernel borra todas las variables  

La mejor práctica es ejecutar las celdas de arriba a abajo después de hacer cambios y reiniciar el kernel.

# Paso 9: Escribir un Archivo en Google Drive

Ahora que tu Google Drive está montado, los mismos comandos que has usado antes para leer y escribir archivos funcionarán aquí también.

La diferencia clave es **dónde** guardas tu trabajo. Si guardas un archivo en tu carpeta de Google Drive montada, permanecerá allí incluso si el entorno de ejecución del cuaderno se desconecta o reinicia.


````python
import random
random.seed(42)
numero_aleatorio = random.randint(100000, 999999)

# Escribir solo en el archivo txt
with open("numero_aleatorio1.txt", "w") as archivo:
    archivo.write(str(numero_aleatorio))
random.seed(567)
with open("numero_aleatorio2.txt", "w") as archivo:
    archivo.write(str(numero_aleatorio))
````

Ahora confirma que el archivo existe:

````python
!ls
````

Los dos archivos escritos arriba serán importantes para completar las preguntas de demostración. Asegúrate de navegar a tu carpeta de Google Drive y confirmar que los archivos existen.

# Paso 11: Crear Accesos Directos de Archivos para Guardar

Para organizar mejor tus archivos y garantizar un acceso fácil, es una buena práctica crear accesos directos en tu Google Drive a carpetas de proyectos específicos. De esta manera, no siempre tienes que navegar por toda la ruta `/content/drive/MyDrive/`.

### Cómo Crear un Acceso Directo
1. **Ve a Google Drive en tu navegador web.**
2. **Navega a la carpeta que quieres usar para tu proyecto.**
3. **Haz clic derecho en la carpeta.**
4. **Selecciona `Añadir acceso directo a Drive`**. Esto creará un acceso directo en tu sección 'Mi unidad'.

Una vez que tengas un acceso directo, puedes navegar fácilmente hacia él desde Colab. Por ejemplo, si creaste un acceso directo llamado `Proyecto` que apunta a tu carpeta de proyecto real, podrías usar:

```python
%cd /content/drive/MyDrive/Proyecto
```

Esto hace mucho más simple guardar y cargar archivos directamente en tu ubicación específica del proyecto.
