## What Is a Jupyter Notebook?

**Hint:** It is what you just downloaded and uploaded.

A Jupyter Notebook is an interactive environment where you can write and run Python code alongside text, explanations, and results in the same document.

When used with Google Colab, a Jupyter notebook lives in your Google Drive and gives you access to a Python environment from almost anywhere.

## Notebook Cells
A Jupyter Notebook is made up of cells. There are two main types of cells you will use.

### Code Cells
- Used to write and run Python code  
- Executed using the Run button or Shift + Enter  
- Output appears directly below the cell  

### Formatted Text Cells
- Used to write text, explanations, or notes  
- Do not execute code  
- Useful for documenting what your code does.

**This cell and all others that look like it are Formatted Text Cells** You can add cells by clicking +Code or +Text above.

## Execution Order Matters ‼️

Jupyter Notebooks do not automatically run from top to bottom. Cells run only when you execute them.

This means:
- Variables can exist even if their cell appears later in the notebook  
- Running cells out of order can cause errors or confusing results  
- Restarting the kernel clears all variables  

Best practice is to run cells from top to bottom after making changes and restarting the kernel.

# Step 11: Creating File Shortcuts for Saving

To better organize your files and ensure easy access, it's a good practice to create shortcuts in your Google Drive to specific project folders. This way, you don't always have to navigate through the entire `/content/drive/MyDrive/` path.

### How to Create a Shortcut:
1.  **Go to Google Drive in your web browser.**
2.  **Navigate to the folder you want to use for your project.** (e.g., a folder named `CS_Project_2`).
3.  **Right-click on the folder.**
4.  **Select `Add shortcut to Drive`**. This will create a shortcut in your 'My Drive' section.

Once you have a shortcut, you can easily `cd` into it from Colab. For example, if you created a shortcut named `CS_Project_2` pointing to your actual project folder, you could then use:

```python
%cd /content/drive/MyDrive/CS_Project_2
```

This makes it much simpler to save and load files directly to your project-specific location.
