### **¿Por Qué Necesitamos la Función `print`?**
La función `print()` en Python es esencial para mostrar datos en la consola. Juega un papel vital en la depuración al permitir a los desarrolladores rastrear valores de variables y el flujo del programa en tiempo real. Además, proporciona una forma simple de mostrar resultados y dar retroalimentación inmediata durante el desarrollo, lo cual es increíblemente útil para solucionar problemas y aprender.

---

### **Conceptos Básicos sobre la Sintaxis de `print`**
La función `print()` es fundamental para mostrar información en Python. Puede aceptar múltiples objetos, convirtiéndolos a cadenas e imprimiéndolos. Por defecto, separa los objetos con un espacio (`sep=' '`) y termina la salida con una nueva línea (`end='\n'`). Aquí hay algunos ejemplos básicos:

- `print("Hello, World!")`: Muestra una cadena simple.
- `print("Hello", "World", sep="-")`: Separador personalizado (guion `-` entre palabras).
- `print("Hello", "World", end=".")`: Carácter de fin personalizado (imprime un punto en lugar de una nueva línea).

---

### **`print` Formateado**
Los literales de cadena formateados, o f-strings, permiten cadenas dinámicas que pueden incluir valores de variables directamente. Creas un f-string prefijando la cadena con `f` o `F`, e insertando expresiones dentro de llaves `{}`. Estas expresiones son evaluadas y reemplazadas con sus valores. Ejemplo:
```python
# ejemplo 1
name = "Mundo"
print(f"Hola, {name}!")  # Salida: Hola, Mundo!

# ejemplo 2
a = 5
b = 10
print(f"Cinco más diez es {a + b}")  # Salida: Cinco más diez es 15
```
