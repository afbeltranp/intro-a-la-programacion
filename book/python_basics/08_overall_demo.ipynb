### **¬øC√≥mo Maneja Python los Tipos de Datos?**  
Python **convierte autom√°ticamente** ciertos tipos de datos (coerci√≥n de tipos) para evitar errores.  

#### **Ejemplo: Divisi√≥n de Enteros vs. Divisi√≥n de Flotantes**  
```python
int_division = 5 // 2   # Divisi√≥n entera (divisi√≥n de piso)
float_division = 5 / 2.0  # Divisi√≥n de flotantes
print(f"{int_division}, {float_division}")  # Salida: 2, 2.5
```
---

### **Conversi√≥n de Tipos en Python**  
La conversi√≥n de tipos te permite **convertir datos** de un tipo a otro usando funciones integradas:  

| **Funci√≥n** | **Descripci√≥n** | **Ejemplo** |
|-------------|----------------|-------------|
| `int()`  | Convierte a entero | `int(10.98) ‚Üí 10` |
| `float()`  | Convierte a flotante | `float(5) ‚Üí 5.0` |
| `str()`  | Convierte a cadena | `str(42) ‚Üí '42'` |
| `bool()`  | Convierte a booleano | `bool(0) ‚Üí False`, `bool(1) ‚Üí True` |

---

### **Ejemplos de Conversi√≥n de Tipos**  

#### **Convertir Flotante a Entero**  
```python
num = 10.98
print("Entero:", int(num))  # Salida: 10
```

#### **Manejar Entrada del Usuario (Conversi√≥n de Entrada)**  
```python
decimal_num = input("Ingresa un flotante: ")  # La entrada siempre es una cadena
decimal_num = int(decimal_num)  # Convertir a entero
print("Entero:", decimal_num)
```

#### **Conversi√≥n de Entrada Abreviada**  
```python
decimal_num = int(input("Ingresa un flotante: "))  
print("Entero:", decimal_num)
```

---

### **Manejar Entrada en Python**  
- La funci√≥n `input()` **siempre devuelve una cadena**, por lo que debes **convertirla** para operaciones num√©ricas.  
- Para operaciones de texto, convertir la entrada a min√∫sculas puede **prevenir problemas de sensibilidad a may√∫sculas**.  

#### **Ejemplo: Convertir Entrada para Operaciones Num√©ricas**  
```python
user_input = input("Por favor ingresa un n√∫mero: ")
number = int(user_input)  # Convertir entrada a entero
print(f"Entrada convertida a entero para c√°lculos: {number + 10}")
```

#### **Ejemplo: Convertir Entrada para Operaciones de Texto**  
```python
user_input = input("Por favor ingresa una palabra: ")
text = user_input.lower()  # Convertir a min√∫sculas
print(f"Entrada convertida a min√∫sculas para operaciones de texto: {text}")
```

---

### **Puntos Clave**  
- Python soporta **cuatro tipos de datos simples primarios**: `int`, `float`, `str` y `bool`.  
- La **coerci√≥n de tipos** ocurre autom√°ticamente en algunos casos, pero la **conversi√≥n de tipos** es necesaria para la entrada del usuario.  
- Usa `int()`, `float()`, `str()` y `bool()` para **convertir tipos de datos** cuando sea necesario.
